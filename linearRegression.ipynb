```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data
```

```python
X = data.iloc[:, 0]
X
```

```python
Y = data.iloc[:, 1]
Y
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
# Y = mX + c
m = 0
c = 0

L = 0.0001 #learning rate

n = float(len(X))

for i in range(1000): 
    Y_pred = m*X + c
    D_m = (-2/n) * sum(X * (Y - Y_pred)) #derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)
    m = m - L * D_m
    c = c - L * D_c

print(m, c)
```

```python
Y_pred = m * X + c
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color = "red")
plt.show()
```
